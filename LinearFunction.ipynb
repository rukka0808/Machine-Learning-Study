Linear Function

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
A_Speed = 1 #@param {type:"slider", min:0, max:10, step: 1}
A_Bias = 4 #@param {type:"slider", min:0, max:10, step: 1}
B_Speed = 2 #@param {type:"slider", min:0, max:10, step: 1}

MAXVAL =10
INTERVAL = (MAXVAL*10) + 1
doMeet = False
A_xdata, A_ydata, B_xdata, B_ydata = [], [], [], []
```

```python
# size
plt.figure(figsize=(10, 10))

for t in np.linspace(0, MAXVAL, INTERVAL):
  A_y = A_Speed*t + A_Bias
  B_y = B_Speed*t
  A_xdata.append(t)
  A_ydata.append(A_y)
  B_xdata.append(t)
  B_ydata.append(B_y)
  if(B_y >= A_y and (not doMeet)): 
    doMeet = True
    meetTime = t
    meetDistance = A_y

plt.plot(A_xdata, A_ydata, label='A')
plt.plot(B_xdata, B_ydata, label='B')

if (doMeet):
    plt.title('B overcomed from '+str(math.ceil(meetTime*100)/100)+'hour(s), '+str(math.ceil(meetDistance*100)/100)+'km(s)', fontsize=16)
    plt.plot(meetTime, meetDistance, 'ro')
else:
    plt.title('They will not meet', fontsize=16)
```
